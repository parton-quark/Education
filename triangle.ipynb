```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
# Undirected Graph
G = nx.Graph()
# Add nodes to the empty graph
G.add_nodes_from([0,1,2,3,4])
# Add edges to the graph
G.add_edge(0,1)

G.add_edge(1,2)
G.add_edge(1,4)

G.add_edge(2,0)
G.add_edge(2,3)
G.add_edge(2,4)

G.add_edge(3,4)

# draw
nx.draw_networkx(G)
plt.show()
```

```python
# Adjacency matrix of the graph
N = nx.to_numpy_matrix(G)
print(N)
```

```python
# Number of triangles in the graph
N = np.trace(N*N*N)
Tri = N/6
```
